# oxyethylene

```python
oxy = [v for v in SMILES if 'OCC' in v or 'CCO' in v]
print(len(oxy))
print(oxy)
```

# polyether

```python
ether = [p for p in SMILES if 'COC' in p.upper()]
```

```python
print(len(ether))
print(ether)
```

# polyphosphazene

```python
phosphazene = [p for p in SMILES if "N=P" in p or "P=N" in p]
print(len(phosphazene))

similar = []
for v in phosphazene:
    if len(v) <= 23:
        similar.append(v)

print(len(similar))
print(similar)
```

# Flourine containing

```python
flouro = [p for p in SMILES if 'F' in p]
print(len(flouro))

flo = [v for v in flouro if len(v) <= 70 and 'N=P' not in v and 'Si' not in v and 'c' not in v]
print(len(flo))
print(flo)
```

# polysiloxane

```python
siloxane = [p for p in SMILES if 'O[Si]' in p or '[Si]O' in p]
print(len(siloxane))

silo = [v for v in siloxane]
print(len(silo))
print(silo)
```
