```python
import numpy as np
import pandas as pd
```

```python
contents = pd.read_csv("potentialpolymers.csv", index_col='Unnamed: 0')
```

```python
SMILES = contents["SMILES"]
```

# polyether

```python
ether = [p for p in SMILES if 'COC' in p.upper()]
```

```python
print(len(ether))
print(ether)
```

# polyphosphazene

```python
phosphazene = [p for p in SMILES if "N=P" in p or "P=N" in p]
print(len(phosphazene))
```

# polysiloxane

```python
siloxane = [p for p in SMILES if 'O[Si]' in p or '[Si]O' in p]
print(len(siloxane))
```

# polyester

```python
ester = [p for p in SMILES if 'C(=O)O' in p.upper() or 'O(O=)C' in p.upper()]
```

```python
print(len(ester))
print(ester)
```

# polycarbonate

```python
carbonate = [p for p in SMILES if 'C(O)O' in p.upper() or 'O(O)C' in p.upper()]
```

```python
print(len(carbonate))
print(carbonate)
```

```python
PH = [p for p in SMILES if '[PH]' in p or '[HP]' in p]
print(len(PH))
```

```python
PH
```

# polyethylene oxide groups

```python
oxy = [p for p in SMILES if 'CCO' in p or 'OCC' in p]
print(f'ethylene oxide {len(oxy)}')
```

```python
non_oxy = set(SMILES).difference(set(oxy))
print(non_oxy)
```

# heterocycles and aromatic 

```python
heterocycles = [p for p in SMILES if 'n' in p]
```

```python
print(len(heterocycles))
heterocycles
```
