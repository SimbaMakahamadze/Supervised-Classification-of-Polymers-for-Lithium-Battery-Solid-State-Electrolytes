```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Each line is a metric; accuracy, precision, recall, F1

models = ['random forest', 'knn', 'neural network', 'svm', 'decision tree',]

bars = {
    'accuracy': [0.6884, 0.6153, 0.7499, 0.7461, 0.6669,],
    'precision': [0.7444, 0.6025, 0.7958, 0.7820, 0.7046,],
    'recall': [0.6884, 0.6153, 0.7499, 0.7461, 0.6669,],
    'f1': [0.6852, 0.6053, 0.7468, 0.7442, 0.6622,],
}

areas = [0.7037, 0.5750, 0.7604, 0.7549, 0.6703,]
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = 1.5 * np.arange(5)
fig = plt.figure(figsize=(16, 8))
ax = fig.add_axes([0,0,1,1])


font1 = {'family':'arial','color':'#868686','size':20}
ax.tick_params(direction='out', length=6, width=2, colors='r',
               grid_color='r', grid_alpha=0.5, labelsize=16)

ax.bar(X + 0.00, bars["accuracy"], color = '#bc5090', width = 0.20, align='center')
ax.bar(X + 0.25, bars["precision"], color = '#58508d', width = 0.20, align='center')
ax.bar(X + 0.50, bars["recall"], color = '#ff6361', width = 0.20, align='center')
ax.bar(X + 0.75, bars["f1"], color = '#ffa600', width = 0.20, align='center')
ax.bar(X + 1.0, areas, color = '#79ccb3', width = 0.20, align='center')

ax.set_xticks(X, labels=models,)
ax.legend(labels=['accuracy', 'precision', 'recall', 'f1-score', 'auroc'])
plt.axhline(y=0.5,linewidth=1, color='k', linestyle='--')

plt.title("Model performance", fontdict=font1)
plt.ylabel("Performance", fontdict=font1)
plt.xlabel("Model", fontdict=font1)

plt.show()

```
