```python
df = pd.read_csv("PI1M.csv")
```

```python
contents = map(int, contents)
```

```python
contents = list(contents)
```

```python
contents
```

```python
contents.sort()
```

```python
recommendation_indexes = np.array(list(contents))
```

```python
recommendations = df.loc[recommendation_indexes]
```

```python
recommendations.to_csv("recommendations.csv")
```

```python
recomms = recommendations['SMILES'].to_list()
```

```python
recomms
```

```python
heteroatoms = ['N', 'O', 'S', 'P', 'Si']
```

```python
def filter_heteroatoms(SMILES):
    for h in heteroatoms:
        if h in SMILES:
            return True
    return False
```

```python
heteroatom_convention = filter(filter_heteroatoms, recomms)
```

```python
len(list(heteroatom_convention))
```

```python
oxygen_filter = lambda y: 'O' in y
```

```python
peo_legacy = filter(oxygen_filter, recomms)
```

```python
len(list(peo_legacy))
```

```python
new_insight_filter = lambda y: 'P' in y and 'N' in y
```

```python
nitrogen_phosphorus = filter(new_insight_filter, recomms)
```

```python
len(list(nitrogen_phosphorus))
```
